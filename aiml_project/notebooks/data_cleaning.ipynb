```python
Q1=df['MonthlyIncome'].quantile(0.25)
Q3=df['MonthlyIncome'].quantile(0.75)

IQR=Q3-Q1

lower=Q1-1.5*IQR
upper=Q3+1.5*IQR

outliers=df[(df['MonthlyIncome']<lower) | (df['MonthlyIncome']>upper)]

outliers
```

```python
df=df[(df['MonthlyIncome']>=lower) & (df['MonthlyIncome']<=upper)]
```

```python
df.columns
```

```python
df = df.rename(columns={
    'BusinessTravel': 'Travel',
    'Department': 'Dept',
    'DistanceFromHome': 'HomeDist',
    'EducationField': 'EduField',
    'EmployeeCount': 'EmpCount',
    'EmployeeNumber': 'EmpNo',
    'JobLevel': 'JobLvl',
    'JobRole': 'JobRole',
    'MaritalStatus': 'Marital',
    'MonthlyIncome': 'MonthlyInc',
    'MonthlyRate': 'MonthlyRate',
    'PercentSalaryHike': 'SalaryHike',
    'StandardHours': 'StdHours',
    'WorkLifeBalance': 'Work_Life_Bal',
})
```
