```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["figure.figsize"] = (8,5) #to set the size of default graph
sns.set_style("whitegrid")
```

```python
df = pd.read_csv("../data/hr.csv")
```

```python
df['Attrition'].value_counts()
```

```python
sns.countplot(x='Attrition', data=df)
plt.title("Employee Attrition")
plt.show()
```

```python
sns.histplot(df['Age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df['MonthlyInc'], bins=25, kde=True)
plt.title("Monthly Income Distribution")
plt.show()
```

```python
sns.histplot(df['TotalWorkingYears'], bins=20, kde=True)
plt.show()
```

```python
sns.histplot(df['YearsAtCompany'], bins=20, kde=True)
plt.show()
```

```python
sns.countplot(x='Gender', data=df)
plt.show()
```

```python
sns.countplot(y='Dept', data=df)
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(y='JobRole', data=df)

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.countplot(y='EduField', data=df)

plt.show()
```

```python
sns.countplot(x='BusinessTravel', data=df)
plt.show()
```

```python
sns.countplot(x='OverTime', data=df)
plt.show()
```

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)

plt.show()
```

```python
sns.boxplot(x='Attrition', y='YearsAtCompany', data=df)

plt.show()
```

```python
numeric_df = df.select_dtypes(include=np.number)
```

```python
plt.figure(figsize=(15,10))

sns.heatmap(numeric_df.corr(),
            annot=True,
            cmap='Pastel2',
            fmt=".2f")

plt.show()
```

```python
plt.figure(figsize=(9,6))

sns.scatterplot(x='Age',
                y='MonthlyInc',
                hue='Attrition',
                data=df)

plt.title("Age vs Monthly Income by Attrition")

plt.show()
```

```python
joblevel_income = df.groupby('JobLvl')['MonthlyInc'].mean().reset_index()

plt.figure(figsize=(8,5))

sns.lineplot(x='JobLvl',
             y='MonthlyInc',
             data=joblevel_income,
             marker='o')

plt.title("Average Monthly Income by Job Level")
plt.xlabel("Job Level")
plt.ylabel("Average Monthly Income")

plt.show()
```

```python
plt.figure(figsize=(15,6))

plt.plot(df.index,
         df['MonthlyInc'],
         marker='.',
         linewidth=1)

plt.title("Monthly Salary of Each Employee")
plt.xlabel("Employee Index")
plt.ylabel("Monthly Income")

plt.show()
```
